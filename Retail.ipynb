### Bivariate Analysis

```python
fig, ax = plt.subplots()
colors = {'Consumer':'red', 'Corporate':'blue', 'Home Office':'green'}
ax.scatter(df['Sales'], df['Profit'], c=df['Segment'].apply(lambda x: colors[x]))
plt.show()
```

```python
df.pivot_table(values='Sales', index='Segment', columns='Discount', aggfunc='median')
```
